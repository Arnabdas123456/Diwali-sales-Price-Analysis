```python
from matplotlib import pyplot as plt 
import pandas as pd 
import numpy as np 
import seaborn as sns
```

```python

#importing Diwali sales csv file 
df=pd.read_csv('Diwali Sales Data.csv',encoding='unicode_escape')
df
```

```python
#checking how many row or columns are inside
df.shape
```

```python
# print first 5 columns
df.head()
```

```python
# checking all columns types
df.info()
```

Drop blank Columns

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True)
df.head()
```

```python
# Checking Null Values 
df.isnull().sum()
```

```python
#drop Null Values
df.dropna(inplace=True)
df.info()
```

```python
df.shape
```

```python
df['Amount']
```

```python
df.columns
```

```python
# Change Data Type

df['Amount']=df['Amount'].astype('int')
df['Amount']
```

```python
# Describe 3 columns 
df[['Age','Orders','Amount']].describe()
```

Exploratory Data Analysis

Gender

```python
# Counting Gender
cn=sns.countplot(x='Gender',data=df,palette='plasma')
for bars in cn.containers:
    cn.bar_label(bars)
plt.show()
```

```python
sales=df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Gender',y='Amount',data=sales,palette='Dark2')
plt.show()
```

Age

```python
# Counting between Age and Age_group
a=sns.countplot(x='Age Group',data=df,hue='Gender',palette='rocket')
for bars in a.containers:
    a.bar_label(bars)
plt.show()
```

```python
# Total amount vs Age group

sales_age=df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sa=sns.barplot(x='Age Group',y='Amount',data=sales_age,palette='mako')
plt.show()
```

State

```python
#Total number of orders from 15 states
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(15)
sns.set(rc={'figure.figsize':(25,10)})
sns.barplot(data=sales_state,x='State',y='Orders',palette='magma')
plt.show()
```

```python
#Total Amount of Sales from 15 states

sales_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(15)

sns.set(rc={'figure.figsize':(25,10)})
sns.barplot(data=sales_state,x='State',y='Amount',palette='crest')

plt.show()
```

Marital Status

```python
# checking Marital Status
ax=sns.countplot(x='Marital_Status',data=df,palette='rocket')
sns.set(rc={'figure.figsize':(7,5)})
for bar in ax.containers:
    ax.bar_label(bar)
plt.show()
```

```python
## Total amount vs Marital_satatus
sales_state=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(x='Marital_Status',y='Amount',data=sales_state,hue='Gender',color='r')
plt.show()
```

Occupation

```python
# checking most Occupation
sns.set(rc={'figure.figsize':(20,5)})
ax=sns.countplot(x='Occupation',data=df,palette='mako')
for bar in ax.containers:
    ax.bar_label(bar)
plt.show()
```

```python
# Total amount vs Occupation
sales_state=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Occupation',y='Amount',data=sales_state,palette='flare')
plt.show()
```

Product Category

```python
# checking which products sales is high
sns.set(rc={'figure.figsize':(25,5)})
ax=sns.countplot(x='Product_Category',data=df,palette='mako')
for bar in ax.containers:
    ax.bar_label(bar)
plt.show()
```

```python
#Total Amount of Product_Category first 10 columns
sales_state=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(30,10)})
sns.barplot(x='Product_Category',y='Amount',data=sales_state,palette='flare')
plt.show()
```
